# 수고(m) & 착과량(int)

```python
fig = plt.figure(figsize=(10, 8)) 
plt.subplot(221)
sns.pointplot(data = train, x = '수고(m)', y = '착과량(int)')
plt.subplot(222)
sns.countplot(data = train, x = '수고(m)')
plt.subplots_adjust(wspace = 2)
plt.tight_layout()
plt.show()
```
